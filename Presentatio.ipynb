## 2.3 Importando o dataset
Esse dataframe é apenas o inicial, ao longo da apresentação, mais dataframes serão criados.

```python
Games_df = pd.read_csv('Datasets\Video_Games.csv')
```

## 2.4 Limpeza e organização dos dados

```python
display(Games_df)
```

### 2.4.1 Limpando o dataset principal

```python
# Verificando valores nulos
display(Games_df.isnull().sum())
display(Games_df.info())
```

```python
# Limpando valores nulos das colunas 'Name', 'Year_of_Release', 'Genre', 'Publisher'
Games_df = Games_df.dropna(subset = ['Name', 'Year_of_Release', 'Genre', 'Publisher'])

# Removendo a coluna 'Developer', que é redundante
Games_df = Games_df.drop('Developer', axis=1)

# Alterando o type da coluna 'Year_of_Release' para inteiro
Games_df['Year_of_Release'] = Games_df['Year_of_Release'].astype('int64')
```

```python
display(Games_df)
```

### 2.4.2 Dividindo o dataset principal

```python
# Data Frame 'GSales' para análise de vendas
GSales_df = Games_df[['Name', 'Platform', 'Year_of_Release', 'Genre', 'Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]

# Data Frame para comparação de Score com Sales
GFinal_df = Games_df.copy()
```

```python
display(GSales_df)
display(GFinal_df)
```

### 2.4.3 Limpando os Data Frames

```python
# Verificando valores nulos
display(GSales_df.isnull().sum())
display(GFinal_df.isnull().sum())
```

```python
# O Data Frame GSales está pronto para ser utilizado
display(GSales_df)
display(GSales_df.info())
```

```python
# Removendo valores nulos do Data Frame GFinal
GFinal_df = GFinal_df.dropna(subset = ['Critic_Score', 'Critic_Count', 'User_Score', 'User_Count', 'Rating'])

# Alterando o type da coluna 'User_Score' para float
GFinal_df['User_Score'] = GFinal_df['User_Score'].apply(lambda x: str(x).replace(",","."))
GFinal_df['User_Score'] = GFinal_df['User_Score'].astype('float64')

# Alterando o type das colunas 'Critic_Count' e 'User_Count' para inteiro
GFinal_df['Critic_Count'] = GFinal_df['Critic_Count'].astype('int64')
GFinal_df['User_Count'] = GFinal_df['User_Count'].astype('int64')

# Ajustando o valor na coluna 'Critic_Score'
GFinal_df['Critic_Score'] = GFinal_df['Critic_Score']/10
```

```python
# Data Frame GFinal pronto para ser utilizado
display(GFinal_df)
display(GFinal_df.info())
```

### 2.5.2 Dados da Nintendo

```python
NintendoSales_df = GSales_df.loc[GSales_df['Publisher'] == 'Nintendo']
NintendoFinal_df = GFinal_df.loc[GFinal_df['Publisher'] == 'Nintendo']
```

```python
display(NintendoSales_df)
display(NintendoSales_df.describe())
```

Nessa pequena descrição obtemos as seguintes informações:
<p>
    &emsp;- O jogo mais antigo da Nintendo, registrado neste Data Frame, foi lançado em 1983, e o mais recente, 2016.
    <br><br>
    &emsp;- A média de vendas dos jogos da Nintendo, lançados entre as datas registradas, é de aproximadamente 1.166 milhões na América do Norte, 0.598 milhões na Europa, 0.654 milhões no Japão e 0.135 milhões no resto do mundo. Totalizando uma média de aproximadamente 2.552 milhões de vendas globais por jogo.
    <br><br>
    Vamos visualizar esses dados.
</p>

```python
NintendoSales_df.boxplot(column=['NA_Sales', 'EU_Sales'])
plt.show()
```

```python
NintendoSales_df.boxplot(column=['JP_Sales', 'Other_Sales'])
plt.show()
```

```python
NintendoSales_df.boxplot(column='Global_Sales')
plt.show()
```

A visualização desses dados por boxplot nos mostra que muitos jogos vendidos pela Nintendo são outliers (extremos) acima do valor máximo, pois fazem um sucesso muito maior em comparação a maioria dos jogos, de forma a deixar o gráfico de certa forma ilegível.

```python
# Analisando o total de vendas dos jogos lançados em cada ano
Nintendo_Year_Sales = NintendoSales_df[['Year_of_Release', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].groupby('Year_of_Release').sum()
```

```python
display(Nintendo_Year_Sales)
```

```python
Nintendo_Year_Sales_80s = Nintendo_Year_Sales.head(7)
```

```python
sns.barplot(data=Nintendo_Year_Sales_80s, x="Year_of_Release", y="Global_Sales")
plt.show()
```

```python
Nintendo_Year_Sales_90s = Nintendo_Year_Sales.drop([1983, 1984, 1985, 1986, 1987, 1988, 1989])
```

```python
sns.barplot(data=Nintendo_Year_Sales_90s.head(10), x="Year_of_Release", y="Global_Sales")
plt.show()
```

```python
Nintendo_Year_Sales_2000s = Nintendo_Year_Sales_90s.drop([1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999])
```

```python
sns.barplot(data=Nintendo_Year_Sales_2000s.head(10), x="Year_of_Release", y="Global_Sales")
plt.show()
```

```python
Nintendo_Year_Sales_2010s = Nintendo_Year_Sales_2000s.drop([2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009])
```

```python
sns.barplot(data=Nintendo_Year_Sales_2010s, x="Year_of_Release", y="Global_Sales")
plt.show()
```

Com esses graficos, podemos observar que os jogos de maior sucesso da Nintendo, em termos de venda, foram lançados em 2006 e os de menor, em 1983. A empresa viveu muitos altos e baixos em relação aos seus jogos e, nos últimos anos analisados, esteve em decadência, desconsiderando os dados de 2016, pois eram lançamentos muito recentes para se obter um número considerável de vendas.

```python
# Analisando vendas por plataforma
Nintendo_Platform_Sales = NintendoSales_df[['Platform', 'Global_Sales']].groupby('Platform').sum()
display(Nintendo_Platform_Sales)
```

```python
sns.barplot(data=Nintendo_Platform_Sales, x="Platform", y="Global_Sales")
plt.show()
```

A plataforma que mais teve sucesso de vendas foi o Nintendo Wii, logo em seguida o Nintendo DS.
