## 2.3 Importando o dataset

Esse dataframe é apenas o inicial, ao longo da apresentação, mais dataframes serão criados.

```python
Games_df = pd.read_csv('Datasets\Video_Games.csv')
```

## 2.4 Limpeza e organização dos dados

```python
display(Games_df)
```

### 2.4.1 Limpando o dataset principal

```python
# Verificando valores nulos
display(Games_df.isnull().sum())
display(Games_df.info())
```

```python
# Limpando valores nulos das colunas 'Name', 'Year_of_Release', 'Genre', 'Publisher'
Games_df = Games_df.dropna(subset = ['Name', 'Year_of_Release', 'Genre', 'Publisher'])

# Removendo a coluna 'Developer', que é redundante
Games_df = Games_df.drop('Developer', axis=1)

# Alterando o type da coluna 'Year_of_Release' para inteiro
Games_df['Year_of_Release'] = Games_df['Year_of_Release'].astype('int64')
```

```python
display(Games_df)
```

### 2.4.2 Dividindo o dataset principal

```python
# Data Frame 'GSales' para análise de vendas
GSales_df = Games_df[['Name', 'Platform', 'Year_of_Release', 'Genre', 'Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]

# Data Frame para comparação de Score com Sales
GScore_df = Games_df.copy()
```

```python
display(GSales_df)
display(GScore_df)
```

### 2.4.3 Limpando os Data Frames

```python
# Verificando valores nulos
display(GSales_df.isnull().sum())
display(GScore_df.isnull().sum())
```

```python
# O Data Frame GSales está pronto para ser utilizado
display(GSales_df)
display(GSales_df.info())
```

```python
# Removendo valores nulos do Data Frame GFinal e colunas que não são necessárias
GScore_df = GScore_df.dropna(subset = ['Critic_Score', 'Critic_Count', 'User_Score', 'User_Count', 'Rating'])
GScore_df = GScore_df.drop(['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], axis=1)

# Alterando o type da coluna 'User_Score' para float
GScore_df['User_Score'] = GScore_df['User_Score'].apply(lambda x: str(x).replace(",","."))
GScore_df['User_Score'] = GScore_df['User_Score'].astype('float64')

# Alterando o type das colunas 'Critic_Count' e 'User_Count' para inteiro
GScore_df['Critic_Count'] = GScore_df['Critic_Count'].astype('int64')
GScore_df['User_Count'] = GScore_df['User_Count'].astype('int64')

# Ajustando o valor na coluna 'Critic_Score'
GScore_df['Critic_Score'] = GScore_df['Critic_Score']/10
```

```python
# Data Frame GFinal pronto para ser utilizado
display(GScore_df)
display(GScore_df.info())
```

## 2.5 Visualização dos dados

Primeiramente, vamos entender quais informações nós temos.

```python
display(GSales_df['Publisher'].nunique(), GScore_df['Publisher'].nunique())
```

Nosso Data Frame GSales possui dados de 579 empresas de jogos, já o Data Frame e GFinal possui dados de 262 empresas diferentes. Como esses valores são muito grandes para serem analisandos de forma conjunta, vamos escolher arbitrariamente aqueles que mais nos interessam.

```python
display(GSales_df['Publisher'].value_counts(), GScore_df['Publisher'].value_counts())
```

```python
# Ordenando os valores de venda global por empresa
GMostSales = GSales_df[['Publisher', 'Global_Sales']].groupby('Publisher').sum().sort_values('Global_Sales', ascending=False)
display(GMostSales)
```
