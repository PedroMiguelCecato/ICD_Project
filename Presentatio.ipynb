## 2.3 Importando o dataset
Esse dataframe é apenas o inicial, ao longo da apresentação, mais dataframes serão criados.

```python
Games_df = pd.read_csv('Datasets\Video_Games.csv')
```

## 2.4 Limpeza e organização dos dados

```python
display(Games_df)
```

### 2.4.1 Limpando o dataset principal

```python
# Verificando valores nulos
display(Games_df.isnull().sum())
display(Games_df.info())
```

```python
# Limpando valores nulos das colunas 'Name', 'Year_of_Release', 'Genre', 'Publisher'
Games_df = Games_df.dropna(subset = ['Name', 'Year_of_Release', 'Genre', 'Publisher'])

# Removendo a coluna 'Developer', que é redundante
Games_df = Games_df.drop('Developer', axis=1)

# Alterando o type da coluna 'Year_of_Release' para inteiro
Games_df['Year_of_Release'] = Games_df['Year_of_Release'].astype('int64')
```

```python
display(Games_df)
```

### 2.4.2 Dividindo o dataset principal

```python
# Data Frame 'GSales' para análise de vendas
GSales_df = Games_df[['Name', 'Platform', 'Year_of_Release', 'Genre', 'Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]

# Data Frame para comparação de Score com Sales
GScore_df = Games_df.copy()
```

```python
display(GSales_df)
display(GScore_df)
```

### 2.4.3 Limpando os Data Frames

```python
# Verificando valores nulos
display(GSales_df.isnull().sum())
display(GScore_df.isnull().sum())
```

```python
# O Data Frame GSales está pronto para ser utilizado
display(GSales_df)
display(GSales_df.info())
```

```python
# Removendo valores nulos do Data Frame GFinal e colunas que não são necessárias
GScore_df = GScore_df.dropna(subset = ['Critic_Score', 'Critic_Count', 'User_Score', 'User_Count', 'Rating'])
GScore_df = GScore_df.drop(['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], axis=1)

# Alterando o type da coluna 'User_Score' para float
GScore_df['User_Score'] = GScore_df['User_Score'].apply(lambda x: str(x).replace(",","."))
GScore_df['User_Score'] = GScore_df['User_Score'].astype('float64')

# Alterando o type das colunas 'Critic_Count' e 'User_Count' para inteiro
GScore_df['Critic_Count'] = GScore_df['Critic_Count'].astype('int64')
GScore_df['User_Count'] = GScore_df['User_Count'].astype('int64')

# Ajustando o valor na coluna 'Critic_Score'
GScore_df['Critic_Score'] = GScore_df['Critic_Score']/10
```

```python
# Data Frame GFinal pronto para ser utilizado
display(GScore_df)
display(GScore_df.info())
```

## 2.5 Visualização dos dados

Primeiramente, vamos entender quais informações nós temos.

### 2.5.2 Dados da Nintendo

```python
# Filtrando
NintendoSales_df = GSales_df.loc[GSales_df['Publisher'] == 'Nintendo']
NintendoScore_df = GScore_df.loc[GScore_df['Publisher'] == 'Nintendo']
```

```python
display(NintendoSales_df)
display(NintendoSales_df.describe())
```

Nessa pequena descrição obtemos as seguintes informações:
<p>
    &emsp;O jogo mais antigo da Nintendo, registrado neste Data Frame, foi lançado em 1983, e o mais recente, 2016.
    <br><br>
    &emsp;A média de vendas dos jogos da Nintendo, lançados entre as datas registradas, é de aproximadamente:
    <br>
    &emsp;1.166 milhões na América do Norte; 
    <br>
    &emsp;598 mil na Europa; 
    <br>
    &emsp;654 mil no Japão; 
    <br>
    &emsp;135 mil no resto do mundo.
    <br>
    &emsp;Totalizando uma média de aproximadamente 2.552 milhões de vendas globais por jogo.
    <br><br>
    &emsp;Até 25% dos jogos tiveram 290 mil vendas globais, ou menos (1º quartil); 
    <br>
    &emsp;Até 50% dos jogos tiveram 900 mil vendas globais, ou menos (esse valor é a mediana);
    <br>
    &emsp;Até 75% dos jogos tiveram 2.2625 milhões de vendas globais, ou menos (3º quartil);
    <br><br>
    &emsp;O jogo que menos vendeu teve 10 mil vendas globais, já o jogo mais vendido teve 82.53 milhões de vendas globais.
    <br><br>
    Vamos visualizar esses dados.
</p>

```python
# Fazendo uso de boxplot
NintendoSales_df.boxplot(column=['NA_Sales', 'EU_Sales'])
plt.show()

NintendoSales_df.boxplot(column=['JP_Sales', 'Other_Sales'])
plt.show()

NintendoSales_df.boxplot(column='Global_Sales')
plt.show()
```

A visualização desses dados por boxplot nos mostra que muitos jogos vendidos pela Nintendo são outliers (extremos) acima do valor máximo, pois fazem um sucesso muito maior em comparação a maioria dos jogos, de forma a deixar o gráfico de certa forma ilegível.

```python
# Analisando o total de vendas dos jogos lançados em cada ano
Nintendo_Year_Sales = NintendoSales_df[['Year_of_Release', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].groupby('Year_of_Release').sum()
display(Nintendo_Year_Sales)

sns.barplot(data=Nintendo_Year_Sales, x="Year_of_Release", y="Global_Sales")
plt.show()

sns.lineplot(data=Nintendo_Year_Sales, x="Year_of_Release", y="Global_Sales")
plt.show()
```

```python
# Ordenando os valores
Nintendo_MYSales = Nintendo_Year_Sales.sort_values('Global_Sales', ascending=False)
display(Nintendo_MYSales)
```

Com esses graficos, podemos observar que os jogos de maior sucesso da Nintendo, em termos de venda, foram lançados em 2006 e os de menor, em 1983.

```python
# Analisando vendas por plataforma
Nintendo_Platform_Sales = NintendoSales_df[['Platform', 'Global_Sales']].groupby('Platform').sum()
display(Nintendo_Platform_Sales)

sns.barplot(data=Nintendo_Platform_Sales, x="Platform", y="Global_Sales")
plt.show()
```

```python
# Analisando vendas por gênero
Nintendo_Genre_Sales = NintendoSales_df[['Genre', 'Global_Sales']].groupby('Genre').sum()
display(Nintendo_Genre_Sales)

sns.barplot(data=Nintendo_Genre_Sales, x='Genre', y='Global_Sales')
plt.show()
```

```python
# Analisando os jogos mais vendidos da Nintendo
Nintendo_MSales = NintendoSales_df.sort_values('Global_Sales', ascending=False)
display(Nintendo_MSales)

sns.barplot(data=Nintendo_MSales.head(10), x = 'Name', y = 'Global_Sales')
plt.show()
```

```python
display(NintendoScore_df)
display(NintendoScore_df.describe())
```

```python
# Analisando o score dos críticos
Nintendo_BScoreC = NintendoScore_df.sort_values('Critic_Score', ascending=False)
display(Nintendo_BScoreC)

sns.barplot(data= Nintendo_BScoreC.head(), x = 'Name', y = 'Critic_Score')
plt.show()

# Comparando com as vendas
sns.barplot(data= Nintendo_BScoreC.head(), x = 'Name', y = 'Global_Sales')
plt.show()
```

```python
# Analisando o score dos usuários
Nintendo_BScoreU = NintendoScore_df.sort_values('User_Score', ascending=False)
display(Nintendo_BScoreU)

sns.barplot(data= Nintendo_BScoreU.head(), x = 'Name', y = 'User_Score')
plt.show()

# Comparando com as vendas
sns.barplot(data= Nintendo_BScoreU.head(), x = 'Name', y = 'Global_Sales')
plt.show()
```
